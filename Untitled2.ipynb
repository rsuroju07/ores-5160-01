```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('/content/lung_disease dataset.csv')
```

```python
data.shape
```

```python
data.info()
```

```python
#EXISTING COLOUMS
data.columns
```

```python
#DROPPING THE UNWANTED COLOUMS
data.drop(columns = ['Scan'], axis=1, inplace=True)
```

```python
#NEW COLOUMS
data.columns
```

```python
###CHANGING NAMES OF THE COLOUMS
data = data.rename(columns = {'Patient': 'PATIENT',
                              'Risk': 'RISK',
                              'smoke': 'SMOKE',
                              'O2': 'OXYGEN',
                              'ABG-P-O2': 'PO2',
                              'ABG-P-CO2': 'PCO2',
                              'ABG-pH Level': 'PH LEVEL',
                              'Asthama': 'ASTHAMA',
                              'Other diseaes': 'OTHER DISEASES'})
data
```

```python
##DROPPING THE DUPLICATES IF THERE ARE ANY
data.drop_duplicates()
```

```python
##REMOVING NULL IF EXISTED
data.dropna()
```

```python
#OUTLIERS FOR FVC
sns.boxplot(data['FVC'])
```

```python
upper_limit = data['FVC'].mean() + 3*data['FVC'].std()
lower_limit = data['FVC'].mean() - 3*data['FVC'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
##FINDING OUTLIERS OF FVC
data.loc[(data['FVC'] > upper_limit) | (data['FVC'] < lower_limit)]
```

```python
##DELETING OUTLIERS OF FVC
new_data1 = data.loc[(data['FVC'] < upper_limit) & (data['FVC'] > lower_limit)]
print('before removing outliers:', len(data))
print('after removing outliers:', len(new_data1))
print('outliers:', len(data)-len(new_data1))
```

```python
#NEW BOX PLOTS
sns.boxplot(new_data1['FVC'])
```

```python
#OUTLIERS OF FEC1
sns.boxplot(data['FEC1'])
```

```python
upper_limit = data['FEC1'].mean() + 3*data['FEC1'].std()
lower_limit = data['FEC1'].mean() - 3*data['FEC1'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
##FINDING OUTLIERS OF FEC1
data.loc[(data['FEC1'] > upper_limit) | (data['FEC1'] < lower_limit)]
```

```python
##DELETING OUTLIERS OF FEC1
new_data2 = data.loc[(data['FEC1'] < upper_limit) & (data['FEC1'] > lower_limit)]
print('before removing outliers:', len(data))
print('after removing outliers:', len(new_data2))
print('outliers:', len(data)-len(new_data2))
```

```python
#NEW BOX PLOT OF FEC1
sns.boxplot(new_data2['FEC1'])
```

```python
#OUTLIERS FOR AGE
sns.boxplot(data['AGE'])
```

```python
upper_limit = data['AGE'].mean() + 3*data['AGE'].std()
lower_limit = data['AGE'].mean() - 3*data['AGE'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
##FINDING OUTLIERS OF AGE
data.loc[(data['AGE'] > upper_limit) | (data['AGE'] < lower_limit)]
```

```python
##DELETING OUTLIERS OF AGE
new_data3 = data.loc[(data['AGE'] < upper_limit) & (data['AGE'] > lower_limit)]
print('before removing outliers:', len(data))
print('after removing outliers:', len(new_data3))
print('outliers:', len(data)-len(new_data3))
```

```python
#NEW BOX PLOT FOR AGE
sns.boxplot(new_data3['AGE'])
```

```python
# CALCULATED FEILDS (FVC AND FEC1)
data = data.assign(FVCFEC1_ratio=data['FVC'] / data['FEC1'])
data.head()
```

```python
#EXISTING DATA TYPES
data.dtypes
```

```python
#CHANGING THE DATA TYPE
columns_to_convert = ['SMOKE', 'PEFR', 'OXYGEN', 'PO2', 'PCO2', 'PH LEVEL', 'ASTHAMA', 'OTHER DISEASES', 'AGE', 'RISK']
data[columns_to_convert] = data[columns_to_convert].replace({'T': True, 'F': False})
data[columns_to_convert] = data[columns_to_convert].astype(bool)
#CHANGED DATA TYPES
data.dtypes
```
